```python
import pandas as pd
import numpy as np
import json
import ast

pd.set_option("display.max_columns", None)
```

```python
with open("../data/raw/games.json", "r", encoding="utf-8") as f:
    steam_data = json.load(f)
```

```python
steam = pd.DataFrame.from_dict(
    steam_data,
    orient="index"
)
```

```python
steam.index.name = "appid"

steam = steam.reset_index()
```

```python
steam.head()
```

```python
steam.info()
```

```python
steam.columns = (
    steam.columns
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
steam.columns
```

```python
remove_columns = [
    "header_image",
    "website",
    "support_url",
    "support_email",
    "detailed_description",
    "about_the_game",
    "short_description",
    "metacritic_url",
    "screenshots",
    "movies",
    "notes"
]

steam = steam.drop(
    columns=remove_columns,
    errors="ignore"
)
```

```python
steam["release_date"].head()
```

```python
steam["release_date"] = pd.to_datetime(
    steam["release_date"],
    errors="coerce"
)
```

```python
steam["release_year"] = (
    steam["release_date"]
    .dt.year
)
```

```python
steam = steam[
    (steam["release_year"] >= 2018) &
    (steam["release_year"] <= 2024)
]
```

```python
steam["price"].describe()
```

```python
steam[
    steam["price"] > 200
]
```

```python
steam.loc[
    steam["price"] > 200,
    "price"
] = np.nan
```

```python
def owner_midpoint(value):
    if pd.isna(value):
        return np.nan
    
    value = value.replace(",", "")
    
    lower, upper = value.split("-")
    
    return (
        int(lower.strip()) +
        int(upper.strip())
    ) / 2
```

```python
steam["estimated_players"] = (
    steam["estimated_owners"]
    .apply(owner_midpoint)
)
```

```python
steam["total_reviews"] = (
    steam["positive"]
    +
    steam["negative"]
)
```

```python
steam["review_score"] = np.where(
    steam["total_reviews"] > 0,
    steam["positive"] /
    steam["total_reviews"],
    np.nan
)
```

```python
steam["genres"] = steam["genres"].apply(
    lambda x: x if isinstance(x, list)
    else []
)
```

```python
steam.info()
```

```python
steam.isna().sum()
```

```python
missing = pd.DataFrame({
    "missing_count": steam.isna().sum(),
    "missing_percentage": (
        steam.isna().sum() / len(steam) * 100
    )
})

missing.sort_values(
    "missing_percentage",
    ascending=False
)
```

```python
steam.to_csv(
    "../data/cleaned/steam_cleaned.csv",
    index=False
)
```
